## **Concepts**
### **Nature of a variable**

Based on the nature of the variable, a different visualization approach was applied.
*   Numeric
*   Categorical

### **Univariate Vs Multivariate grouping**

Here, the types of visualizations that can be performed based on this classification, organizing them into two segments:

* **Univariate:**
  *   **Numeric**:

      Histogram, Scatter plot, Box Plot, KDE, Line chart
  *   **Categorical**:

      Bar Plot, Pie Chart

*   **Multivariate**
    * **Numeric vs Numeric**:

      Scatter plot, Line chart, Heat map
    * **Numeric vs Categorical**:

      Bar Plot, Box Plot
    * **Categorical vs Categorical**:

      Bar Plot, Pie Chart, Stacked bar plot

For each type of visualization, I applied the concept of Graph Summary Components (**"BeCAI"** Frame work), which consists of two main stages:

1.   **Base Plotting**
2.   **Enhancement**
  
  The Enhancement stage is further divided into three layers:


1.   **Contextual (or Conceptual) Layer:**
  In this layer, necessary transformations are applied to the data before plotting.
  ***Examples***: log transformation, arithmetic operations on variables and similar techniques
2.   **Aesthetic Layer:**
This layer focuses on improving the visual appearance of the plot.
Key elements include:

      *   Color
      *   Size
      *   Shape

      These help in differentiating and highlighting data points effectively.

3.   **Information Layer:**
This layer improves the clarity of the plot by adding key elements such as titles, subtitles, themes, templates, axis labels and more.

This structured approach helped in creating effective visualizations to present the data clearly.

# Base of Histogram plotting
**Default** — bin count, blue bars, "count" y-label, grey background, column name as axis label, hover tooltip

**Not default** — title, subtitle, units, reference line, source note, bargap, bar borders, white background



```python
px.histogram(df, x="age",)
```


## **Enhancement on Histogram**

### **Contextual Layer**

* Mean value annotated using python function of `mean()` which can not be calculated in plotly by its own

### **Aesthetic Layer**

* White theme with Arial font
* Clean bars with black borders
* Balanced figure size and margins

### **Informative Layer**

* Bold centred title with subtitle
* X-axis and Y-axis are labeled
* Hover shows exact values


```python
# ── i) CONTEXTUAL LAYER ───────────────────────────────────
fig = px.histogram(df, x="age")

overall_mean = df["age"].mean()
fig.add_vline(
    x                    = overall_mean,
    line_dash            = "dash",
    line_color           = "#888",
    line_width           = 1,
    annotation_text      = f"Mean age = {overall_mean:.1f}",
    annotation_position  = "top right",
    annotation_font_size = 10,
)

# ── ii) AESTHETIC LAYER ───────────────────────────────────
fig.update_traces(
    marker = dict(
        color   = "White",
        opacity = 0.8,
        line    = dict(color="Black", width=1),
    ),
)

# ── iii) INFORMATIVE LAYER ────────────────────────────────
fig.update_layout(
    font          = dict(family="Arial, sans-serif", size=13),
    plot_bgcolor  = "white",
    paper_bgcolor = "white",
    width         = 750,
    height        = 500,
    margin        = dict(l=70, r=30, t=90, b=70),

    title = dict(
        text     = "<b>Distribution of Customer Age</b>",
        subtitle = dict(
            text = "  One variable: Age",
            font = dict(size=11, color="grey"),
        ),
        x        = 0.5,
        xanchor  = "center",
        font     = dict(size=17, color="#111"),
    ),

    xaxis = dict(
        title     = dict(text="Age (years)", font=dict(size=13)),
        showgrid  = False,
        zeroline  = False,
        showline  = True,
        linecolor = "black",
        linewidth = 1,
    ),

    yaxis = dict(
        title     = dict(text="Count", font=dict(size=13)),
        showgrid  = True,
        gridcolor = "rgba(0,0,0,0.06)",
        zeroline  = False,
    ),


    hovermode  = "closest",
    showlegend = False,
)

fig.show()
```

## Enhancement on Pie Chart

### Contextual Layer

* Counted the frequency of each job category using python function of `value_counts()` and converted the index into columns and reset it to the default integer index using `reset_index()` for including count in **hover**.

### Aesthetic Layer

* Colorblind-safe palette used
* White borders added to separate slices
* Clean font and white background
* Wider figure size to fit legend


### Informative Layer

* "Admin" slice pulled slightly to highlight it
* Centered title with a short subtitle
* Right-side vertical legend with border
* Pre-calculated counts used for accurate sizing and order


```python
# ── i) CONTEXTUAL LAYER ───────────────────────────────────
#This helps to include counts in the hover
job_counts = df["job"].value_counts().reset_index()
job_counts.columns = ["job", "count"]

fig = px.pie(job_counts, names="job", values="count")
#fig= px.pie(df, names="job")
# ── ii) AESTHETIC LAYER ───────────────────────────────────
fig.update_traces(
    marker = dict(
        colors = px.colors.qualitative.Safe,
        line   = dict(color="white", width=1.5),
    ),
    #pull = [0.1 if j == "admin." else 0 for j in df["job"]],
    pull = [0.1 if j == "admin." else 0 for j in job_counts["job"]]
)

# ── iii) INFORMATIVE LAYER ────────────────────────────────
fig.update_layout(
    font          = dict(family="Arial, sans-serif", size=13),
    plot_bgcolor  = "white",
    paper_bgcolor = "white",
    width         = 700,
    height        = 500,
    margin        = dict(l=70, r=30, t=90, b=70),


    title = dict(
        text     = "<b>Distribution of Job Categories</b>",
        subtitle = dict(
            text = " One variable: Job  ·  Admin slice pulled",
            font = dict(size=11, color="grey"),
        ),
        x        = 0.5,
        xanchor  = "center",
        font     = dict(size=17, color="#111"),
    ),

    legend = dict(
        title       = dict(text="Job"),
        orientation = "v",
        x           = 1.02,
        y           = 1,
        bordercolor = "rgba(0,0,0,0.1)",
        borderwidth = 1,
    ),
)


fig.show()
```

```python
# ── i) CONTEXTUAL LAYER ───────────────────────────────────
job_marital = df.groupby(["job", "marital"]).size().reset_index(name="count")

fig = px.bar(job_marital, x="job", y="count", color="marital", barmode="group")

# ── ii) AESTHETIC LAYER ───────────────────────────────────
fig.update_traces(
    marker = dict(
        opacity = 0.85,
        line    = dict(color="white", width=0.5),
    ),
)

fig.update_layout(
    colorway = ["#1a5276", "#2e86c1", "#a9cce3"],
)

# ── iii) INFORMATIVE LAYER ────────────────────────────────
fig.update_layout(
    font          = dict(family="Arial, sans-serif", size=13),
    plot_bgcolor  = "white",
    paper_bgcolor = "white",
    width         = 850,
    height        = 500,
    margin        = dict(l=70, r=30, t=90, b=100),

    title = dict(
        text     = "<b>Job vs Marital Status ( Count)</b>",
        subtitle = dict(
            text = "n = Number of customers  ·   Job × Marital",
            font = dict(size=11, color="grey"),
        ),
        x        = 0.5,
        xanchor  = "center",
        font     = dict(size=17, color="#111"),
    ),

    xaxis = dict(
        title     = dict(text="Job", font=dict(size=13)),
        showgrid  = False,
        zeroline  = False,
        tickangle = 45,
    ),

    yaxis = dict(
        title     = dict(text="Count", font=dict(size=13)),
        showgrid  = True,
        gridcolor = "rgba(0,0,0,0.06)",
        zeroline  = False,
    ),

    hovermode  = "closest",
    legend     = dict(
        title       = dict(text="Marital"),
        orientation = "v",
        x           = 1.02,
        y           = 1,
        bordercolor = "rgba(0,0,0,0.1)",
        borderwidth = 1,
    ),
)


fig.show()
```

## **Enhancement of Grid**

### **Contextual Layer**

* Counts created using Python functions `value_count()` which is plotly can not calculate it by its own.


### **Aesthetic Layer**

* Colorblind-safe palette (Set2) applied
* Large figure height for better readability
* Cleaned grid labels (removed prefixes)

### **Informative Layer**

* Title shows all three variables (Age × Job × Marital)
* Facet grid created using job (rows) and marital (columns)
* Independent y-axes for each subplot
* Tick labels enabled on all panels for readability



```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Count values for each column
job_counts   = df['job'].value_counts()
month_counts = df['month'].value_counts()
loan_counts  = df['loan'].value_counts()

fig = make_subplots(
    rows=2, cols=3,
    subplot_titles=(
        'Job ', 'Month ', 'Loan',
        'Job ',  'Month ',  'Loan '
    ),
    specs=[
        [{"type": "bar"}, {"type": "bar"}, {"type": "bar"}],
        [{"type": "pie"}, {"type": "pie"}, {"type": "pie"}]
    ]
)

# ── Row 1: Bar charts ──────────────────────────────────────
fig.add_trace(go.Bar(x=job_counts.index,   y=job_counts.values,   name='Job',   showlegend=False), row=1, col=1)
fig.add_trace(go.Bar(x=month_counts.index, y=month_counts.values, name='Month', showlegend=False), row=1, col=2)
fig.add_trace(go.Bar(x=loan_counts.index,  y=loan_counts.values,  name='Loan',  showlegend=False), row=1, col=3)

# ── Row 2: Pie charts ──────────────────────────────────────
fig.add_trace(go.Pie(labels=job_counts.index,   values=job_counts.values,   name='Job',   hole=0.3), row=2, col=1)
fig.add_trace(go.Pie(labels=month_counts.index, values=month_counts.values, name='Month', hole=0.3), row=2, col=2)
fig.add_trace(go.Pie(labels=loan_counts.index,  values=loan_counts.values,  name='Loan',  hole=0.3), row=2, col=3)

fig.update_layout(
    title_text="Univariate Analysis — Job, Month, Loan",
    height=750,
    showlegend=False
)

fig.show()
```

```python

# ════════════════════════════════════════════════════════════════
# CONTEXTUAL LAYER
# ════════════════════════════════════════════════════════════════

df['age_group'] = pd.cut(df['age'],
                          bins=[18, 30, 40, 50, 60, 90],
                          labels=['18-30', '31-40', '41-50', '51-60', '60+'])

grouped = df.groupby(['age_group', 'job', 'marital'], observed=True).size().reset_index(name='count')

age_groups    = ['18-30', '31-40', '41-50', '51-60', '60+']
marital_cats  = grouped['marital'].unique().tolist()
palette       = px.colors.qualitative.Set2
color_map     = {m: palette[i] for i, m in enumerate(marital_cats)}

# Row 1: 3 panels of equal width  → colspan=2 each in a 6-col grid
# Row 2: 2 panels, each wider     → colspan=3 each in a 6-col grid
specs = [
    [{"colspan": 2}, None, {"colspan": 2}, None, {"colspan": 2}, None],
    [{"colspan": 3}, None, None, {"colspan": 3}, None, None],
]

fig = make_subplots(
    rows=2, cols=6,
    specs=specs,
    subplot_titles=[g for g in age_groups],
    horizontal_spacing=0.06,
    vertical_spacing=0.18,
)

panel_positions = {
    '18-30': (1, 1),
    '31-40': (1, 3),
    '41-50': (1, 5),
    '51-60': (2, 1),
    '60+':   (2, 4),
}

shown_in_legend = set()

for age_grp, (row, col) in panel_positions.items():
    sub = grouped[grouped['age_group'] == age_grp]

    for marital in marital_cats:
        sub_m       = sub[sub['marital'] == marital]
        show_legend = marital not in shown_in_legend

        fig.add_trace(
            go.Bar(
                x           = sub_m['job'],
                y           = sub_m['count'],
                name        = marital,
                marker_color= color_map[marital],
                legendgroup = marital,
                showlegend  = show_legend,
            ),
            row=row, col=col,
        )
        shown_in_legend.add(marital)


fig.add_annotation(
    xref      = "paper", yref="paper",
    x=1, y=-0.08, showarrow=False,
    text      = "Source: Bank Marketing Dataset",
    xanchor   = "right",
    font      = dict(size=10, color="grey"),
)

# ════════════════════════════════════════════════════════════════
# AESTHETIC LAYER
# ════════════════════════════════════════════════════════════════

fig.update_layout(
    template  = "simple_white",
    barmode   = "group",
    height    = 700,
    legend    = dict(title="Marital status", x=1.01, y=0.5),
)

fig.update_xaxes(tickangle=45)

# ════════════════════════════════════════════════════════════════
# INFORMATIVE LAYER
# ════════════════════════════════════════════════════════════════

fig.update_layout(
    title = dict(
        text     = "Age Group × Job × Marital Status",
        x        = 0.5,
        xanchor  = "center",
        font     = dict(size=16),
    ),
)

fig.update_yaxes(title_text="Count")

fig.show()
```


## **Enhancement on Facet Bar Plot**

### **Contextual Layer**

* Counts created using Python functions `groupby()`, `size()`, and `reset_index()` which can not be calculated in plotly by its own.


### **Aesthetic Layer**

* Colorblind-safe palette (Set2) applied
* Large figure height for better readability
* Cleaned facet labels
* X-axis labels rotated for clarity


### **Informative Layer**

* Title shows all three variables (Age × Job × Marital)
* Facet grid created using job (rows) and marital (columns)
* Independent y-axes (dynamic) for each subplot
* Tick labels enabled on all panels for readability


```python
import plotly.express as px

fig = px.bar(
    df.groupby(['age', 'job', 'marital']).size().reset_index(name='count'),
    x='age',
    y='count',
    facet_col='marital',
    facet_row='job',
    color='job',
    title='Age × Job × Marital Status',
    height=1800,
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.update_xaxes(tickangle=45)

# ── MAKE EACH SUBPLOT'S Y-AXIS INDEPENDENT ──────────────────────
fig.update_yaxes(matches=None)          # breaks the shared y-axis linkage
fig.update_yaxes(showticklabels=True)   # ensures tick labels show on every panel

fig.show()
```
