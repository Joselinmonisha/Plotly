# **Data structure and BeCAI**

### **Nature of a variable**

Based on the nature of the variable, a different visualization approach was applied.
*   Numeric
*   Categorical

### **Univariate Vs Multivariate grouping**

Here, the types of visualizations that can be performed based on this classification, organizing them into two segments:

* **Univariate:**
  *   **Numeric**:

      Histogram, Scatter plot, Box Plot, KDE, Line chart
  *   **Categorical**:

      Bar Plot, Pie Chart

*   **Multivariate**
    * **Numeric vs Numeric**:

      Scatter plot, Line chart
    * **Numeric vs Categorical**:

      Box Plot, Violin
    * **Categorical vs Categorical**:

      Grouped Bar Plot, Stacked bar plot

# **BeCAI framework**

For each type of visualization, I applied the concept of Graph Summary Components (**"BeCAI"** Frame work), which consists of two main stages:

1.   **Base Plotting**
2.   **Enhancement**
  
  The Enhancement stage is further divided into three layers:


1.   **Contextual (or Conceptual) Layer:**
  In this layer, necessary transformations are applied to the data before plotting.
  ***Examples***: log transformation, arithmetic operations on variables and similar techniques
2.   **Aesthetic Layer:**
This layer focuses on improving the visual appearance of the plot.
Key elements include:

      *   Color
      *   Size
      *   Shape

      These help in differentiating and highlighting data points effectively.

3.   **Information Layer:**
This layer improves the clarity of the plot by adding key elements such as titles, subtitles, themes, templates, axis labels and more.

This structured approach helped in creating effective visualizations to present the data clearly.
