# Base of Histogram plotting
**Default** — bin count, blue bars, "count" y-label, grey background, column name as axis label, hover tooltip

**Not default** — title, subtitle, units, reference line, source note, bargap, bar borders, white background



```python
px.histogram(df, x="age",)
```


## **Enhancement on Histogram**

### **Contextual Layer**

* Mean value annotated using python function of `mean()` which can not be calculated in plotly by its own

### **Aesthetic Layer**

* White theme with Arial font
* Clean bars with black borders
* Balanced figure size and margins

### **Informative Layer**

* Bold centred title with subtitle
* X-axis and Y-axis are labeled
* Hover shows exact values


```python
# ── i) CONTEXTUAL LAYER ───────────────────────────────────
fig = px.histogram(df, x="age")

overall_mean = df["age"].mean()
fig.add_vline(
    x                    = overall_mean,
    line_dash            = "dash",
    line_color           = "#888",
    line_width           = 1,
    annotation_text      = f"Mean age = {overall_mean:.1f}",
    annotation_position  = "top right",
    annotation_font_size = 10,
)

# ── ii) AESTHETIC LAYER ───────────────────────────────────
fig.update_traces(
    marker = dict(
        color   = "White",
        opacity = 0.8,
        line    = dict(color="Black", width=1),
    ),
)

# ── iii) INFORMATIVE LAYER ────────────────────────────────
fig.update_layout(
    font          = dict(family="Arial, sans-serif", size=13),
    plot_bgcolor  = "white",
    paper_bgcolor = "white",
    width         = 750,
    height        = 500,
    margin        = dict(l=70, r=30, t=90, b=70),

    title = dict(
        text     = "<b>Distribution of Customer Age</b>",
        subtitle = dict(
            text = "  One variable: Age",
            font = dict(size=11, color="grey"),
        ),
        x        = 0.5,
        xanchor  = "center",
        font     = dict(size=17, color="#111"),
    ),

    xaxis = dict(
        title     = dict(text="Age (years)", font=dict(size=13)),
        showgrid  = False,
        zeroline  = False,
        showline  = True,
        linecolor = "black",
        linewidth = 1,
    ),

    yaxis = dict(
        title     = dict(text="Count", font=dict(size=13)),
        showgrid  = True,
        gridcolor = "rgba(0,0,0,0.06)",
        zeroline  = False,
    ),


    hovermode  = "closest",
    showlegend = False,
)

fig.show()
```

# **Base of Box Plot**

**Default** — five-number summary, whiskers, outlier dots, axis lable,  hover tooltip, grey background

**Not default** — title, subtitle, units, reference line, source note, marker styling, box line width, white background, tick rotation, hover



```python
px.box(df, y="balance")
```

## **Enhanced Box Plot**

### **Contextual Layer**

* Direct visualization of balance distribution (no grouping applied)

### **Aesthetic Layer**

* Steel blue color with slight transparency
* box width with clear borders
* Outliers shown as light hollow circles
* Clean white theme with Arial font

### **Informative Layer**

* Centered title with subtitle
* Y-axis labeled and X-axis labels removed (not needed for single variable)
* Subtle horizontal gridlines added


```python

# ── i) CONTEXTUAL LAYER ───────────────────────────────────
fig = px.box(df, y="balance")

# ── ii) AESTHETIC LAYER ───────────────────────────────────
fig.update_traces(
    marker = dict(
        color   = "steelblue",
        size    = 10,
        opacity = 0.6,
        symbol  = "circle-open",
    ),
    width      = 0.35,
    line_width = 1.2,
    boxmean    = True,
    selector   = dict(type="box")
)

# ── iii) INFORMATIVE LAYER ────────────────────────────────
fig.update_layout(
    font          = dict(family="Arial, sans-serif", size=13),
    plot_bgcolor  = "white",
    paper_bgcolor = "white",
    width         = 800,
    height        = 500,
    margin        = dict(l=70, r=30, t=90, b=70),

    title = dict(
        text     = "<b>Distribution of Account Balance</b>",
        subtitle = dict(
            text = "One variable: Balance",
            font = dict(size=11, color="grey"),
        ),
        x        = 0.5,
        xanchor  = "center",
        font     = dict(size=17, color="#111"),
    ),

    xaxis = dict(
        showticklabels = False,
        showgrid       = False,
        zeroline       = False,
    ),

    yaxis = dict(
        title     = dict(text="Account Balance", font=dict(size=13)),
        showgrid  = True,
        gridcolor = "rgba(0,0,0,0.06)",
        zeroline  = False,
    ),

    hovermode  = "closest",
    showlegend = False,
)


fig.show()
```

## Base of Pie Chart
**Default** — slice sizes, percentage labels, default palette, legend, solid circle, hover tooltip

**Not default** — title, subtitle, source note, donut hole, colorblind palette, slice borders, textinfo control, white background



```python
px.pie(df, names="job")
```

## Enhancement on Pie Chart

### Contextual Layer

* Counted the frequency of each job category using python function of `value_counts()` and converted the index into columns and reset it to the default integer index using `reset_index()` for including count in **hover**.

### Aesthetic Layer

* Colorblind-safe palette used
* White borders added to separate slices
* Clean font and white background
* Wider figure size to fit legend


### Informative Layer

* "Admin" slice pulled slightly to highlight it
* Centered title with a short subtitle
* Right-side vertical legend with border
* Pre-calculated counts used for accurate sizing and order


```python
# ── i) CONTEXTUAL LAYER ───────────────────────────────────
#This helps to include counts in the hover
job_counts = df["job"].value_counts().reset_index()
job_counts.columns = ["job", "count"]

fig = px.pie(job_counts, names="job", values="count")
#fig= px.pie(df, names="job")
# ── ii) AESTHETIC LAYER ───────────────────────────────────
fig.update_traces(
    marker = dict(
        colors = px.colors.qualitative.Safe,
        line   = dict(color="white", width=1.5),
    ),
    #pull = [0.1 if j == "admin." else 0 for j in df["job"]],
    pull = [0.1 if j == "admin." else 0 for j in job_counts["job"]]
)

# ── iii) INFORMATIVE LAYER ────────────────────────────────
fig.update_layout(
    font          = dict(family="Arial, sans-serif", size=13),
    plot_bgcolor  = "white",
    paper_bgcolor = "white",
    width         = 700,
    height        = 500,
    margin        = dict(l=70, r=30, t=90, b=70),


    title = dict(
        text     = "<b>Distribution of Job Categories</b>",
        subtitle = dict(
            text = " One variable: Job  ·  Admin slice pulled",
            font = dict(size=11, color="grey"),
        ),
        x        = 0.5,
        xanchor  = "center",
        font     = dict(size=17, color="#111"),
    ),

    legend = dict(
        title       = dict(text="Job"),
        orientation = "v",
        x           = 1.02,
        y           = 1,
        bordercolor = "rgba(0,0,0,0.1)",
        borderwidth = 1,
    ),
)


fig.show()
```

# **Important finding**
In plotly bar plot couldn't calculate count on its own


```python

# ── i) CONTEXTUAL LAYER ───────────────────────────────────
contact_counts = df["contact"].value_counts().reset_index()
contact_counts.columns = ["contact", "count"]

fig = px.bar(contact_counts, x="contact", y="count")

# ── ii) AESTHETIC LAYER ───────────────────────────────────
fig.update_traces(
    marker = dict(
        color   = "steelblue",
        opacity = 0.85,
        line    = dict(color="white", width=0.5),
    ),
)

# ── iii) INFORMATIVE LAYER ────────────────────────────────
fig.update_layout(
    font          = dict(family="Arial, sans-serif", size=13),
    plot_bgcolor  = "white",
    paper_bgcolor = "white",
    width         = 550,
    height        = 500,
    margin        = dict(l=70, r=30, t=90, b=70),

    title = dict(
        text     = "<b>Distribution of Contact Type</b>",
        subtitle = dict(
            text = "  One variable: Contact",
            font = dict(size=11, color="grey"),
        ),
        x        = 0.5,
        xanchor  = "center",
        font     = dict(size=17, color="#111"),
    ),

    xaxis = dict(
        title     = dict(text="Contact Type", font=dict(size=13)),
        showgrid  = False,
        zeroline  = False,
    ),

    yaxis = dict(
        title     = dict(text="Count", font=dict(size=13)),
        showgrid  = True,
        gridcolor = "rgba(0,0,0,0.06)",
        zeroline  = False,
    ),

    hovermode  = "closest",
    showlegend = False,
)



fig.show()
```
